```python
import yfinance as yf
import pandas as pd
import plotly.graph_objects as go
from datetime import datetime
```

```python
start_date = '1996-01-01'
end_date = datetime.today().strftime("%Y-%m-%d")
```

```python
data = yf.download(['USDKRW=X'],start=start_date, end=end_date)
```

```python
data
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(x=data.index,
                         y=data.Open,
                         mode='lines',
                         name='달러환율'))

fig.update_layout(title='달러환율',
                  plot_bgcolor='white')

fig.add_hline(y=1300, line_width=2, line_dash="solid", line_color="red")
fig.add_hline(y=1200, line_width=2, line_dash="solid", line_color="blue")
fig.add_hline(y=1100, line_width=2, line_dash="solid", line_color="green")
fig.update_xaxes(showgrid=True, gridwidth=1, gridcolor='lightgrey')
fig.update_yaxes(showgrid=True, gridwidth=1, gridcolor='lightgrey')

fig.show()
```
