```python
import pandas as pd
import plotly.graph_objects as go
```

```python
# csv 파일 로드
dizel_df = pd.read_csv('data/normal_dizel20220220_20220606.csv', encoding='cp949')
gas_df = pd.read_csv('data/normal_gas20220220_20220606.csv', encoding='cp949')
dizel_df_2012 = pd.read_csv('data/normal_gas20120101_20131231.csv', encoding='cp949')
```

```python
from datetime import datetime
import locale

locale.setlocale(locale.LC_TIME, "ko_KR.UTF-8")

date_time_str = dizel_df.구분[0]
date_time_obj = datetime.strptime(date_time_str, '%Y년%m월%d일')

print('Date-time : ', date_time_obj)
```

```python
date_dizel = pd.to_datetime(dizel_df.구분, format="%Y년%m월%d일")
date_gas = pd.to_datetime(gas_df.구분, format="%Y년%m월%d일")
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(x=date_dizel, y=dizel_df.서울, mode='lines',
                         name='경유 가격',
                         line=dict(color='#7f7f7f')))
fig.add_trace(go.Scatter(x=date_gas, y=gas_df.서울, mode='lines',
                         name='휘발유 가격',
                         line=dict(color='#1f77b4')))

fig.update_layout(title="2022년 1월 20일부터 6월 6일까지 유가 변화",
                  plot_bgcolor="white")

fig.update_xaxes(title_text="날짜")
fig.update_yaxes(title_text="원(₩)",
                 range=[1400,2200])

# shape 추가
fig.add_shape(type="line",
              x0=pd.to_datetime('2022-02-24'), y0=1400,
              x1=pd.to_datetime('2022-02-24'), y1=2200,
              line=dict(dash='dot',
                        color='#CB4C4E'))
# shape에 대한 text
fig.add_trace(go.Scatter(x=[pd.to_datetime('2022-03-03')], y=[2150],
                         text=["우크라이나 침공"],
                         mode="text",
                         name="우크라이나 침공"))

fig.show()
```
